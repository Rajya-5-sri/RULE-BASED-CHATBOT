```python
import re
import random

commodities = {
    "potato": 30,
    "potatoes": 30,
    "tomato": 40,
    "tomatoes": 40,
    "onion": 35,
    "onions": 35,
    "rice": 60,
    "wheat": 45,
    "carrot": 50,
    "beans": 70
}

greetings = ["Hello!", "Hi there!", "Hey!"]
goodbyes = ["Goodbye!", "See you later!"]

def extract_commodity_and_quantity(user_input):
    user_input = user_input.lower()

    qty_match = re.search(r'(\d+(?:\.\d+)?)\s*kg', user_input)
    quantity = float(qty_match.group(1)) if qty_match else 1

    for item in commodities:
        if item in user_input:
            return item, quantity

    return None, quantity

def chatbot():
    while True:
        user_input = input("User: ").lower()

        if any(word in user_input for word in ["hi", "hello", "hey"]):
            print("Bot:", random.choice(greetings))

        elif any(word in user_input for word in ["bye", "exit"]):
            print("Bot:", random.choice(goodbyes))
            break

        elif any(word in user_input for word in ["price", "cost", "rate"]):
            item, quantity = extract_commodity_and_quantity(user_input)

            if item:
                total_cost = commodities[item] * quantity
                print(f"Bot: The cost of {quantity} kg of {item} is ₹{total_cost:.2f}")
            else:
                print("Bot: Sorry, I could not find that commodity in the price list.")

        else:
            print("Bot: I didn't understand that.")
            
chatbot()
```
